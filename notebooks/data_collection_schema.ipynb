1. LLM промптится для генерации естесственного жизненного сценария. При генераци задаются:
    * одна из пяти эмоций, которую испытывает герой
    * пол и возрастная группа героя
    * тип социального взаимодействия, которое вызвало соответсвующую эмоцию в сценарии

    эти параметры задаются для повышения разнообразия сценариев (если не задавать эти параметры, то разнообразия не будет вообще, например не указывая пол и возраст при почти любой адекватной температуре генерировались сйенарии для 19 yo. college student по имени Sarah). Прмер сценария - мужчина средних лет чувствует злость на своего коллегу, который не делает задачи в срок и из-за этого подрывает планы коллег.
2. С этим же промптом LLM генерирует этот сценарий с точки зрения собеседника главного героя. При этом собеседник не знает всех нюансов ситуации, а ему дана лишь часть информации, он тут внешний наблюдатель
3. Эти сценарии валидируются с помощью LLM на согласованность друг с другом и общую "адекватность"
4. Две LLM промптятся для ведения диалога друг с другом. Промпт состоит из соответсвующих сценариев. При этом для собеседника главного героя сценария есть два вида промптов:
   * Эмпатичный - в него передаем дополнительно эмоцию из сценария
   * Неэмпатичный - просим llm иметь низкий эмоциональный интеллект.


Итого для варианта эмпатичных диалогов, собеседник абсолютно точно знает:
* Эмоцию собеседника
* Контекст диалога (в результате чего собеседник испытывает данную эмоцию)

Поэтому, кажется, достаточно провести финальную фильтрацию диалогов (с помощью человеческой разметки или опять с LLM) чтобы получить датасет достаточно хорошего качества, или же можно оставить все как есть, визуально датасет выглядит приемлемо
Фильтровать можно следующим образом:
1. Проверить действительно ли в сгенерированном диалоге человек общается как испытывающий данную в сценарии эмоцию
2. Действительно ли второй собеседник эмпатичен/неэмпатичен
